```python
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import load_model
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    classification_report, confusion_matrix
)

# --- KONFIGURASI PATH MODEL DAN HISTORY ---
train_df = pd.read_csv("D:/ITERA/SEMESTER 8/cnn-modeling-ver2/dataset/train_data_70.csv")
val_df = pd.read_csv("D:/ITERA/SEMESTER 8/cnn-modeling-ver2/dataset/val_data_15.csv")
test_df = pd.read_csv("D:/ITERA/SEMESTER 8/cnn-modeling-ver2/dataset/test_data_15.csv")

# --- Pisahkan fitur dan label ---
X_train = train_df.drop("DEPRESSED", axis=1).values
y_train = train_df["DEPRESSED"].values

X_val = val_df.drop("DEPRESSED", axis=1).values
y_val = val_df["DEPRESSED"].values

X_test = test_df.drop("DEPRESSED", axis=1).values
y_test = test_df["DEPRESSED"].values

# --- Scaling ---
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

# --- Reshape untuk CNN 1D ---
X_train_scaled = X_train_scaled.reshape(X_train_scaled.shape[0], X_train_scaled.shape[1], 1)
X_val_scaled = X_val_scaled.reshape(X_val_scaled.shape[0], X_val_scaled.shape[1], 1)
X_test_scaled = X_test_scaled.reshape(X_test_scaled.shape[0], X_test_scaled.shape[1], 1)
```

```python
import os
import json
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import load_model
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

# --- Folder model dan history (gunakan raw string untuk path Windows) ---
model_dir = r"d:\ITERA\SEMESTER 8\cnn-modeling-ver2\models"
history_dir = r"d:\ITERA\SEMESTER 8\cnn-modeling-ver2\history"

# --- Ambil semua file model ---
model_files = [f for f in os.listdir(model_dir) if f.endswith(".h5")]

for model_file in sorted(model_files):
    config_name = model_file.replace("model_", "").replace(".h5", "")
    
    model_path = os.path.join(model_dir, model_file)
    history_path = os.path.join(history_dir, f"history_{config_name}.json")

    # --- Cek apakah file history ada ---
    if not os.path.exists(history_path):
        print(f"⚠️ History file not found for {config_name}. Skipping...")
        continue

    # --- Load model dan history ---
    model = load_model(model_path)
    with open(history_path, "r") as f:
        history = json.load(f)

    # --- Prediksi ---
    y_pred_prob = model.predict(X_test_scaled)
    y_pred = (y_pred_prob > 0.5).astype("int32")

    # --- Hitung Metrics ---
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)


    # --- Plot Akurasi dan Loss ---
    plt.figure(figsize=(12, 6))
    plt.suptitle(f'Model: {config_name}', fontsize=16)

    # Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history['accuracy'], label='Train Accuracy')
    plt.plot(history['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    # Loss
    plt.subplot(1, 2, 2)
    plt.plot(history['loss'], label='Train Loss')
    plt.plot(history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

    # --- Confusion Matrix ---
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['No Depression', 'Depression'],
                yticklabels=['No Depression', 'Depression'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(f'Confusion Matrix - {config_name}')
    plt.tight_layout()
    plt.show()

    # --- Classification Report ---
    print(f"\n📄 Classification Report - {config_name}")
    print(classification_report(y_test, y_pred, digits=4))

    print(f"\n📊 Evaluasi untuk Model: {config_name}")
    print(f"Accuracy : {acc:.4f}")
    print(f"Precision: {prec:.4f}")
    print(f"Recall   : {rec:.4f}")
    print(f"F1 Score : {f1:.4f}")
```

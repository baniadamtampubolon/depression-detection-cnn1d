```python
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import load_model
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    classification_report, confusion_matrix
)

# --- KONFIGURASI PATH MODEL DAN HISTORY ---
model_paths = {}

for i in range(1, 10):  # Dari Exp-1 hingga Exp-8
    key = f"P{i}"
    model_file = f"model/trainTest{i}_model.h5"
    history_file = f"history/trainTest{i}_history.json"
    model_paths[key] = (model_file, history_file)


train_df = pd.read_csv("dataset/train_data.csv")
val_df = pd.read_csv("dataset/val_data.csv")
test_df = pd.read_csv("dataset/test_data.csv")

# --- Pisahkan fitur dan label ---
X_train = train_df.drop("DEPRESSED", axis=1).values
y_train = train_df["DEPRESSED"].values

X_val = val_df.drop("DEPRESSED", axis=1).values
y_val = val_df["DEPRESSED"].values

X_test = test_df.drop("DEPRESSED", axis=1).values
y_test = test_df["DEPRESSED"].values

# --- Scaling ---
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

# --- Reshape untuk CNN 1D ---
X_train_scaled = X_train_scaled.reshape(X_train_scaled.shape[0], X_train_scaled.shape[1], 1)
X_val_scaled = X_val_scaled.reshape(X_val_scaled.shape[0], X_val_scaled.shape[1], 1)
X_test_scaled = X_test_scaled.reshape(X_test_scaled.shape[0], X_test_scaled.shape[1], 1)

# --- PROSES LOOP SEMUA MODEL ---
for name, (model_file, history_file) in model_paths.items():
    print(f"\nüì¶ Evaluating model: {name}")

    # Load model dan history
    model = load_model(model_file)
    with open(history_file, "r") as f:
        history = json.load(f)

    # Prediksi
    y_pred_prob = model.predict(X_test_scaled)
    y_pred = (y_pred_prob > 0.5).astype("int32")

    # --- Hitung Metrics ---
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    print(f"Accuracy: {acc:.4f} | Precision: {prec:.4f} | Recall: {rec:.4f} | F1 Score: {f1:.4f}")

    # --- Plot Akurasi dan Loss ---
    plt.figure(figsize=(12, 6))
    plt.suptitle(f'Model: {name}', fontsize=16)

    # Accuracy plot
    plt.subplot(1, 2, 1)
    plt.plot(history['accuracy'], label='Train Accuracy')
    plt.plot(history['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    # Loss plot
    plt.subplot(1, 2, 2)
    plt.plot(history['loss'], label='Train Loss')
    plt.plot(history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

    # --- Confusion Matrix (Heatmap) ---
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['No Depression', 'Depression'],
                yticklabels=['No Depression', 'Depression'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(f'Confusion Matrix - {name}')
    plt.tight_layout()
    plt.show()

    # --- Classification Report ---
    print(f"\nüìä Classification Report - {name}")
    print(classification_report(y_test, y_pred, digits=4))

    print(f"Accuracy: {acc:.4f}")
    print(f"Precision: {prec:.4f}")
    print(f"Recall: {rec:.4f}")
    print(f"F1 Score: {f1:.4f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
batch_sizes = [16, 32, 64, 128]
f1_lr_0001 = [90.14, 87.32, 89.66, 90.14]
f1_lr_00001 = [83.58, 86.96, 83.58, 83.58]
f1_lr_00005 = [0, 92.75, 0, 0]

# Bar width
bar_width = 0.25
index = np.arange(len(batch_sizes))

# Create plot
fig, ax = plt.subplots(figsize=(10, 6))

bar_1 = ax.bar(index, f1_lr_0001, bar_width, label='LR = 0.001')
bar_2 = ax.bar(index + bar_width, f1_lr_00001, bar_width, label='LR = 0.0001')
bar_3 = ax.bar(index + 2*bar_width, f1_lr_00005, bar_width, label='LR = 0.0005')

# Add annotations
for bars in [bar_1, bar_2, bar_3]:
    for bar in bars:
        height = bar.get_height()
        if height > 0:
            ax.annotate(f'{height:.2f}',
                        xy=(bar.get_x() + bar.get_width() / 2, height),
                        xytext=(0, 3),
                        textcoords='offset points',
                        ha='center', va='bottom')

# Labels and formatting
ax.set_xlabel('Batch Size')
ax.set_ylabel('F1-Score (%)')
ax.set_title('Perbandingan F1-Score Berdasarkan Batch Size')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(batch_sizes)
ax.set_ylim(80, 100)
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data (lengkapi sesuai hasil F1-score Anda)
models = ['baseline', 'p1', 'p2', 'p3', 'p4', 'p5', 'p6', 'p7', 'p8']
epoch_stop = [30, 16, 20, 25, 30, 50, 50, 50, 50]
f1_scores = [92.75, 90.14, 87.32, 89.66, 90.14, 83.58, 86.96, 83.58, 83.58]  # <-- Gantilah ini dengan data Anda

# Warna berbeda untuk tiap titik
colors = plt.cm.tab10.colors[:len(models)]

# Plot
plt.figure(figsize=(10, 6))
for i in range(len(models)):
    plt.scatter(epoch_stop[i], f1_scores[i], color=colors[i], s=110, label=models[i])  # titik besar
    # Tampilkan nama model dan nilai F1-score
    label_text = f'{models[i]} ({f1_scores[i]:.1f}%)'
    plt.text(epoch_stop[i], f1_scores[i] - 0.6, label_text, fontsize=9, ha='center', va='top')

plt.title('Pengaruh Jumlah Epoch (Early Stop) terhadap F1-Score')
plt.xlabel('Epoch Stop')
plt.ylabel('F1-Score (%)')
plt.ylim(80, 95)
plt.grid(True)
plt.tight_layout()
plt.show()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
batch_sizes = [16, 32, 64, 128]
f1_lr_0001 = [88.33, 91.94, 90.76, 89.08]
f1_lr_00001 = [89.66, 89.83, 90.76, 88.89]


# Bar width
bar_width = 0.25
index = np.arange(len(batch_sizes))

# Create plot
fig, ax = plt.subplots(figsize=(10, 6))

bar_1 = ax.bar(index, f1_lr_0001, bar_width, label='LR = 0.001')
bar_2 = ax.bar(index + bar_width, f1_lr_00001, bar_width, label='LR = 0.0001')


# Add annotations
for bars in [bar_1, bar_2]:
    for bar in bars:
        height = bar.get_height()
        if height > 0:
            ax.annotate(f'{height:.2f}',
                        xy=(bar.get_x() + bar.get_width() / 2, height),
                        xytext=(0, 3),
                        textcoords='offset points',
                        ha='center', va='bottom')

# Labels and formatting
ax.set_xlabel('Batch Size')
ax.set_ylabel('F1-Score (%)')
ax.set_title('Perbandingan F1-Score Berdasarkan Batch Size')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(batch_sizes)
ax.set_ylim(80, 100)
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
batch_sizes = [16, 32, 64, 128]
f1_lr_0001 = [88.33, 91.94, 90.76, 89.08]
f1_lr_00001 = [89.66, 89.83, 90.76, 88.89]

# Bar width
bar_width = 0.25
index = np.arange(len(batch_sizes))

# Create plot
fig, ax = plt.subplots(figsize=(10, 6))

# Bar plots
bar_1 = ax.bar(index, f1_lr_0001, bar_width, label='LR = 0.001')
bar_2 = ax.bar(index + bar_width, f1_lr_00001, bar_width, label='LR = 0.0001')

# Line plots on top of bars
ax.plot(index + bar_width / 2, f1_lr_0001, color='blue', marker='o', linestyle='-', linewidth=1)
ax.plot(index + bar_width * 1.5, f1_lr_00001, color='orange', marker='o', linestyle='-', linewidth=1)


# Add annotations
for bars in [bar_1, bar_2]:
    for bar in bars:
        height = bar.get_height()
        if height > 0:
            ax.annotate(f'{height:.2f}',
                        xy=(bar.get_x() + bar.get_width() / 2, height),
                        xytext=(0, 3),
                        textcoords='offset points',
                        ha='center', va='bottom')

# Labels and formatting
ax.set_xlabel('Batch Size')
ax.set_ylabel('F1-Score (%)')
ax.set_title('Visualisasi Learning Rate terhadap F1-Score')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(batch_sizes)
ax.set_ylim(80, 100)
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data (lengkapi sesuai hasil F1-score Anda)
models = ['p1', 'p2', 'p3', 'p4', 'p5', 'p6 = base', 'p7', 'p8']
epoch_stop = [18, 35, 40, 50, 8, 12, 12, 30]
f1_scores = [89.66, 89.83, 90.76, 88.89, 88.33, 91.94, 90.76, 89.08]  # <-- Gantilah ini dengan data Anda

# Warna berbeda untuk tiap titik
colors = plt.cm.tab10.colors[:len(models)]

# Plot
plt.figure(figsize=(10, 6))
for i in range(len(models)):
    plt.scatter(epoch_stop[i], f1_scores[i], color=colors[i], s=110, label=models[i])  # titik besar
    # Tampilkan nama model dan nilai F1-score
    label_text = f'{models[i]} ({f1_scores[i]:.1f}%)'
    plt.text(epoch_stop[i], f1_scores[i] - 0.6, label_text, fontsize=9, ha='center', va='top')

plt.title('Pengaruh Jumlah Epoch (Early Stop) terhadap F1-Score')
plt.xlabel('Epoch Stop')
plt.ylabel('F1-Score (%)')
plt.ylim(80, 95)
plt.grid(True)
plt.tight_layout()
plt.show()
plt.show()
```
